**Paper:** Baselines and Bigrams: Simple, Good Sentiment and Topic Classification (Wang & Manning, ACL 2012)

## Step-by-Step Method Description

### Step 1: Feature Extraction and Binarization
- Raw text is tokenized and converted to a feature count vector **f⁽ⁱ⁾ ∈ ℝ|V|** for each training example i, where V is the vocabulary.
- These counts are **binarized**: x̂⁽ᵏ⁾ = **1**{f⁽ᵏ⁾ > 0}, replacing raw counts with 0/1 presence indicators.
- **Reference:** Section 2.1, just before the MNB definition.
- **Purpose:** Binarization removes term-frequency bias and is shown empirically to improve both MNB and NBSVM by ~1% over raw counts.

---

### Step 2: Compute NB Log-Count Ratio r (Eq. 2)
- Positive and negative class count vectors are accumulated:
  **p** = α + Σᵢ:y⁽ⁱ⁾=1 f̂⁽ⁱ⁾, **q** = α + Σᵢ:y⁽ⁱ⁾=−1 f̂⁽ⁱ⁾  (α=1 Laplace smoothing)
- The log-count ratio is computed per feature:  
  **r** = log( (p/‖p‖₁) / (q/‖q‖₁) )
- **Reference:** Equation (2), Section 2.
- **Purpose:** r encodes how much more (or less) likely each word is in the positive vs negative class — the core NB discriminative signal.

---

### Step 3a: MNB Prediction (baseline variant)
- Weight vector **w = r**, bias **b = log(N₊/N₋)**.
- Prediction: y⁽ᵏ⁾ = sign(**r**ᵀ x̂⁽ᵏ⁾ + b)
- **Reference:** Section 2.1, Equation (1).
- **Purpose:** This is the Multinomial Naive Bayes classifier in linear form, strong for short snippets.

---

### Step 3b: NBSVM — SVM on NB-Weighted Features (Section 2.3)
- Instead of raw binarized features, the SVM receives **NB-scaled features**:  
  f̃⁽ᵏ⁾ = r̂ ⊙ f̂⁽ᵏ⁾  (element-wise product)
- An L2-regularized, L2-loss linear SVM is trained on these scaled features (Eq. 3) using LIBLINEAR.
- **Reference:** Section 2.3.
- **Purpose:** The NB ratio pre-weights features by their discriminative power, giving the SVM a better-conditioned input space.

---

### Step 4: Interpolation Between MNB and SVM (Eq. 4)
- The final NBSVM weight vector is interpolated:  
  **w'** = (1 − β) w̄ + β **w**  
  where w̄ = ‖w‖₁/|V| is the mean magnitude of w, and β ∈ [0, 1].
- **Reference:** Equation (4), Section 2.3.
- **Purpose:** This acts as a regularizer — trust the NB signal unless the SVM is highly confident. The paper finds β=0.25 optimal for snippets, β∈[¼,1] for long documents.

---

### Step 5: Feature Type Selection (Unigrams vs. Bigrams)
- The entire pipeline (Steps 1–4) is run for both unigram and bigram feature sets.
- **Reference:** Section 4.4 (Table 2, 3, 4).
- **Purpose:** The paper demonstrates that bigrams consistently help sentiment tasks but have mixed utility in topical classification, because sentiment is often expressed in multi-word phrases (e.g., "not bad").

---

## Final Summary Sentence

This paper solves the problem of text sentiment and topic classification by systematically identifying which variants of Naive Bayes and SVM — specifically binarized MNB, an L2-SVM on NB-weighted bigram features with MNB interpolation (NBSVM) — outperform recently published state-of-the-art methods across multiple datasets; the authors claim their advantage comes from carefully combining generative NB statistics with discriminative SVM training, which had not been previously investigated with appropriate feature engineering and model selection care.
