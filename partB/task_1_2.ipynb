# Task 1.2 — Key Assumptions
**Paper:** Baselines and Bigrams (Wang & Manning, ACL 2012)

## Assumption 1

**Assumption:** The NB log-count ratio r is a meaningful and stable feature representation — i.e., features (words/bigrams) that are predictive in the training set remain predictive at test time, and the ratio of class-conditional word frequencies is a reliable discriminative signal.

**Why the method needs it:** NBSVM scales every feature by rⱼ (Eq. 2) before feeding it to the SVM. If r is noisy or not predictive (e.g., because the training set is too small or the domain shifts), then pre-multiplying by r corrupts rather than enhances the feature space, and the SVM sees worse inputs than if it used raw binarized features.

**Violation scenario:** A dataset with severe class imbalance (e.g., 95% negative) where many positive examples share a few rare words — r would be dominated by those rare tokens and would not generalise to an i.i.d. test set. Similarly, cross-domain sentiment (train on product reviews, test on political speeches) would break the assumption because the same words have different class-conditional distributions.

**Paper reference:** Section 2.3 and Equation (2) — the entire NBSVM construction depends on r being a useful feature weight.

---

## Assumption 2

**Assumption:** Binarizing word counts (replacing term frequency with presence/absence) provides a better signal than raw counts for sentiment tasks — i.e., *whether* a sentiment word appears matters more than *how many times* it appears.

**Why the method needs it:** Both MNB and NBSVM use x̂⁽ᵏ⁾ = **1**{f⁽ᵏ⁾ > 0} instead of f⁽ᵏ⁾. The authors report that binarized MNB is "slightly better (by 1%) than using the raw count feature" (Section 4.6). If this assumption fails, the binarization step discards informative frequency information, and the baseline would outperform the proposed model.

**Violation scenario:** Authorship attribution or spam detection tasks where the sheer repetition frequency of specific terms (e.g., a spammer repeating "free" 50 times) is the primary discriminative signal — binarization destroys this information and would hurt performance.

**Paper reference:** Section 2.1 ("we find that binarizing f⁽ᵏ⁾ is better") and Section 4.6.

---

## Assumption 3

**Assumption:** For short snippet tasks (average ~20 words), a bag-of-features (BoF) model captures sufficient information for sentiment — i.e., word order and syntactic structure are not necessary, and sentiment can be inferred from feature co-occurrence alone.

**Why the method needs it:** The entire method family (MNB, SVM, NBSVM) treats documents as bags of features, discarding order. The paper's main empirical claim — that MNB/NBSVM outperform syntactic models like Tree-CRF and RAE on snippet tasks — only holds if BoF features are sufficient for those tasks. If word order or negation scope were critical, a bag model could not compete.

**Violation scenario:** Texts with heavy negation or syntactic inversion (e.g., "not at all bad" → positive; "everything that could be called good is absent" → negative) where a bag-of-features model cannot determine sentiment polarity without understanding scope and syntax.

**Paper reference:** Section 4.2, where the authors discuss snippets like "not an inhumane monster" and "killing cancer" as edge cases; Table 2 compares against Tree-CRF (Nakagawa et al., 2010) and RAE (Socher et al., 2011).
