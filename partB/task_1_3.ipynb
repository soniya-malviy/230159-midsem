# Task 1.3 — What the Paper Claims to Improve
**Paper:** Baselines and Bigrams (Wang & Manning, ACL 2012)

## Main Baseline

The primary baselines compared against are **standard linear SVM with bag-of-words features (BoWSVM)** and **Tree-CRF (Nakagawa et al., 2010)** — a dependency tree-based model using CRFs with hidden variables — as well as **RAE (Recursive Autoencoders, Socher et al., 2011)**.

BoWSVM represents the naive discriminative baseline, while Tree-CRF and RAE represent the sophisticated, structure-sensitive models that were considered state-of-the-art at the time.

---

## Limitation of the Baselines

The paper argues that researchers have been using either (a) a plain linear SVM on bag-of-words features — which is too weak for snippet sentiment — or (b) complex syntactic models like Tree-CRF and RAE — which require parse trees, external lexicons, or unsupervised pretraining. The key limitation identified is that **nobody had carefully optimised the NB/SVM model variants, feature types (bigrams vs. unigrams), and binarization** before declaring these methods outclassed. The paper shows that the apparently inferior baseline family, when properly tuned, matches or exceeds the "advanced" methods.

---

## How the Proposed Method Overcomes This

NBSVM overcomes this by combining the best of both worlds: it uses the discriminative log-count ratio r from NB as a feature-weighting layer before the SVM, and then interpolates the resulting SVM weights with the NB direction (Eq. 4). This hybrid is both computationally simple (no parse tree, no pretraining) and empirically strong across all dataset types.

---

## Condition Where Paper's Method Would NOT Outperform the Baseline

NBSVM would likely not outperform a syntactic baseline under conditions where **fine-grained compositional negation is the primary signal** — for example, a carefully curated dataset of short sentences where sentiment flips depend entirely on negation scope (e.g., "I would not say this is not worth watching" → positive). In this scenario, the bag-of-features assumption (Assumption 3 in Task 1.2) is severely violated: the unigram feature 'not' and 'worth' appear together, but their combination is double-negated positive. A Tree-CRF model that explicitly represents dependency paths between 'not' and 'worth' would assign correct polarity, while NBSVM's log-count ratio r would assign a negative weight to 'not' and a positive weight to 'watching', leading to an ambiguous or wrong prediction.

The paper itself hints at this limitation by noting examples like "not an inhumane monster" (Section 4.2) — exactly the failure mode that rule-based reversal systems were designed to handle and that the paper waves away statistically, but which would cause targeted failures in a negation-focused benchmark.
